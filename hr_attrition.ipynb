**2. Identifying Factors Affecting Attrition**

```python
# Gender
attrition_by_gender = df.groupby('gender').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_gender)
plt.figure()
sns.barplot(x='gender', y='attrition_percentage', data=attrition_by_gender)
plt.tight_layout()
plt.show()
```

```python
# Age Group
def age_group(age):
    if 18 <= age <= 25:
        return '18-25 years old'
    elif 25 < age <= 40:
        return '25-40 years old'
    elif 40 < age <= 60:
        return '40-60 years old'
    else:
        return '>60 years'

df['age_group'] = df['age'].apply(age_group)

attrition_by_age = df.groupby('age_group').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_age)

plt.figure()
sns.barplot(x='age_group', y='attrition_percentage', data=attrition_by_age)
plt.tight_layout()
plt.show()
```

```python
# Job Role
attrition_by_role = df.groupby('job_role').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_role)

attrition_by_role.sort_values(by='attrition_percentage', ascending=False, inplace=True)

plt.figure(figsize=(10,5))
sns.barplot(x='job_role', y='attrition_percentage', data=attrition_by_role)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Job Satisfaction
attrition_by_sat = df.groupby('job_satisfaction').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_sat)

plt.figure()
sns.barplot(x='job_satisfaction', y='attrition_percentage', data=attrition_by_sat)
plt.tight_layout()
plt.show()
```

```python
# Job Involvement
attrition_by_inv = df.groupby('job_involvement').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_inv)

plt.figure()
sns.barplot(x='job_involvement', y='attrition_percentage', data=attrition_by_inv)
plt.tight_layout()
plt.show()
```

```python
# Education
attrition_by_edu = df.groupby('education').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_edu)

plt.figure()
sns.barplot(x='education', y='attrition_percentage', data=attrition_by_edu)
plt.tight_layout()
plt.show()
```

```python
# Income Quartile
df['income_quartile'] = pd.qcut(df['monthly_income'], q=4, labels=[1, 2, 3, 4])

attrition_by_inqu = df.groupby('income_quartile').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_inqu)

attrition_by_inqu.sort_values(by='income_quartile', inplace=True)

plt.figure()
sns.barplot(x='income_quartile', y='attrition_percentage', data=attrition_by_inqu)
plt.tight_layout()
plt.show()
```

```python
# Tenure

def tenure_group(years):
    if 0 <= years <= 2:
        return '0-2 years'
    elif 2 < years <= 5:
        return '2-5 years'
    elif 5 < years <= 8:
        return '5-8 years'
    elif 8 < years <= 10:
        return '8-10 years'
    else:
        return '>10 years'

df['tenure'] = df['years_at_company'].apply(tenure_group)


attrition_by_tenure = df.groupby('tenure').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

attrition_by_tenure = attrition_by_tenure.sort_values(by='tenure')

print(attrition_by_tenure)

plt.figure()
sns.barplot(x='tenure', y='attrition_percentage', data=attrition_by_tenure)
plt.tight_layout()
plt.show()
```

```python
# Professional Experience

def experience_group(years):
    if 0 <= years <= 5:
        return '0-5 years'
    elif 5 < years <= 10:
        return '5-10 years'
    elif 10 < years <= 20:
        return '10-20 years'
    else:
        return '>20 years'

df['experience'] = df['total_working_years'].apply(experience_group)


attrition_by_experience = df.groupby('experience').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

attrition_by_experience = attrition_by_experience.sort_values(by='attrition_percentage', ascending=False)

print(attrition_by_experience)

plt.figure()
sns.barplot(x='experience', y='attrition_percentage', data=attrition_by_experience)
plt.tight_layout()
plt.show()
```

```python
# Environment Satisfaction
attrition_by_env = df.groupby('environment_satisfaction').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_env)

plt.figure()
sns.barplot(x='environment_satisfaction', y='attrition_percentage', data=attrition_by_env)
plt.tight_layout()
plt.show()
```

```python
# Work-Life Balance
# Education
attrition_by_wlb = df.groupby('work_life_balance').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

print(attrition_by_wlb)

plt.figure()
sns.barplot(x='work_life_balance', y='attrition_percentage', data=attrition_by_wlb)
plt.tight_layout()
plt.show()
```

```python
# Living Distance

def distance_group(years):
    if 0 <= years <= 5:
        return '0-5 km'
    elif 5 < years <= 10:
        return '5-10 km'
   
    else:
        return '>10 years'

df['living_distance'] = df['distance_from_home'].apply(distance_group)


attrition_by_distance = df.groupby('living_distance').apply(
    lambda x: round((x[x['attrition'] == 'Yes'].shape[0] * 100.0) / x.shape[0], 2)
).reset_index(name='attrition_percentage')

# attrition_by_distance = attrition_by_distance.sort_values(by='attrition_percentage', ascending=False)

print(attrition_by_distance)

plt.figure()
sns.barplot(x='living_distance', y='attrition_percentage', data=attrition_by_distance)
plt.tight_layout()
plt.show()
```

**3. Exploring Factors Affecting Attrition Rates**

```python
# Age & Income towards Attrition Percentage
result_age = df.groupby('age_group').agg(
    average_income=('monthly_income', 'mean'),
    attrition_percentage=('attrition', lambda x: round((x.eq('Yes').sum() * 100.0) / len(x), 2))
).reset_index()

result_age['average_income'] = result_age['average_income'].round(2)

result_age
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='monthly_income', hue='attrition', data=df, alpha=0.7)

plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.title('Relationship between Age, Monthly Income, and Attrition')

plt.tight_layout()
plt.show()
```

```python
# Job Role & Age towards Attrition Percentage
result_job = df.groupby('job_role').agg(
    average_age=('age', 'mean'),
    attrition_percentage=('attrition', lambda x: round((x.eq('Yes').sum() * 100.0) / len(x), 2))
).reset_index()

result_job['average_age'] = result_job['average_age'].round(2)

result_job
```

```python
plt.figure(figsize=(12, 6))

sns.stripplot(x='job_role', y='age', hue='attrition', data=df, jitter=True, dodge=True)

plt.xlabel('Job Role')
plt.ylabel('Age')
plt.title('Age Distribution by Job Role and Attrition Status')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

```python
# Job Role & Involvement Level towards Attrition Percentage
result_jobinv = df.groupby('job_role').agg(
    average_inv=('job_involvement', 'mean'),
    attrition_percentage=('attrition', lambda x: round((x.eq('Yes').sum() * 100.0) / len(x), 2))
).reset_index()

result_jobinv['average_inv'] = result_jobinv['average_inv'].round(2)

result_jobinv
```

```python
plt.figure(figsize=(12, 6))

# Create a violin plot to visualize job involvement across job roles, colored by attrition
sns.violinplot(x='job_role', y='job_involvement', hue='attrition', data=df, split=True)

# Add labels and title
plt.xlabel('Job Role')
plt.ylabel('Job Involvement (1-5)')
plt.title('Job Involvement by Job Role and Attrition Status')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
```
